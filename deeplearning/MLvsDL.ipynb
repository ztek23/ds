```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
#load the iris dataset 
iris = load_iris()
X = iris.data
print(X)
```

```python
Y = iris.target.reshape(-1, 1)
print(Y)
```

```python
# OneHotEncoding for neural network training
encoder = OneHotEncoder(sparse_output = False)
Y_encoded = encoder.fit_transform(Y)
```

```python
#splitting the data for both models

X_train, X_test, Y_train, Y_test = train_test_split(X, Y_encoded, test_size = 0.2, random_state = 42)

X_train_nn, X_test_nn, Y_train_nn, Y_test_nn = train_test_split(X, Y_encoded, test_size = 0.2, random_state = 42)
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, Y_train)
Y_pred_tree = tree_model.predict(X_test)
tree_accuracy = accuracy_score(Y_test, Y_pred_tree)

print("Decision tree accuracy: ", tree_accuracy)
```

```python
# Deep learning approach neural network
nn_model = Sequential([
    Dense(16, activation = "relu", input_shape = (4, )), 
    Dense(16, activation = "relu"), 
    Dense(3, activation = "softmax")
    ])
```

```python
# compiling the neural network model
# optimizer = adam

nn_model.compile(optimizer = "adam", loss = "categorical_crossentropy", 
                 metrics = ["accuracy"])
nn_model.fit(X_test_nn, Y_test_nn, epochs = 50, batch_size = 5, validation_split = 0.1, verbose = 1)

#evaluating neural network model
nn_loss, nn_accuracy =  nn_model.evaluate(X_test_nn, Y_test_nn, verbose = 0)

print("neural network accuracy", nn_accuracy)
```
